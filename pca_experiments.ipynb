```python
import numpy as np
import matplotlib.pyplot as plt

n=150
t=np.random.randn(n)
noise=1
x=0.1*t + np.random.randn(n)*noise
y=-0.5*t + np.random.randn(n)*noise
plt.scatter(x,y,1,'k')
plt.title("Raw points")
plt.gca().set_aspect(1)
```

```python
plt.scatter(x,y,1,'k')
plt.title("points with variation")

points = np.stack([x,y])
cov = np.cov(points)
basis = np.stack([np.zeros(cov.shape), cov], -1)

# eval, evec = np.linalg.eig(cov)


plt.plot(*basis[0])
plt.plot(*basis[1])

sample = (np.random.rand(2,150)-0.5)*1
plt.scatter(*sample, s=1,c='r')
plt.scatter(*(cov@sample),s=1,c='m')
# plt.plot([0,cov[0,0]], [0,0])
# plt.plot([cov[0,0],cov
# [0,0]], [0,cov[0,1]])
# plt.plot([0,0],[0,cov[1,1]])
# plt.plot([0,cov[1,0]], [cov[1,1],cov[1,1]])



plt.gca().set_aspect(1)
plt.grid(True)
```
